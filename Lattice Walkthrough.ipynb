Build a simple 2 dimensional square lattice

```python
sq_lattice = square((100, 100))
```

```python
sq_model = LattIsing(lattice=sq_lattice)
```

```python
# We save the Matplotlib axes object in case we need to add titles or
# other visual elements
ax = sq_model.plot()
```

```python
sq_model.lattice.lattice.cell_size / 2
```

```python
sq_model.lattice.lattice.superindex_from_pos((1, 1.5))
```

```python
# sq_model.lattice.extend(size=0, ax=0)
```

```python
sq_model.lattice.lattice.limits()
```

```python
sq_model.lattice.properties.shape
```

Now, we split the lattice into a 2x2 grid

```python
sq_model.lattice.build_smc((2, 4))
```

```python
sq_model.lattice.plot_smc_locs()
```

```python
ax = sq_model.plot()
```

```python
sq_model.lattice.lattice.plot(show_indices=True)
```
